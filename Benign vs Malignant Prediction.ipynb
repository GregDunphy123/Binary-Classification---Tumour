## Train Test Split

```python
X = df.drop('diagnosis', axis = 1).values # Our features (numpy array format)
y = df['diagnosis'].values # Our true labels (numpy array format)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# Complete our train test split leaving 20% of data for testing, as quite a small dataset
```

## Models
## 1) Logistic Regression

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
# Fit the model to our split data
```

```python
predictions = logmodel.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test, predictions)) 
# Pass in our true values vs our predictions
```

```python
confusion_matrix(y_test, predictions)
# Shows our true positives/negatives vs false positives/negatives
```

```python
performance = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
performance
# Shows our predicted class (benign or malignant) vs the true class
```

## 3) Neural Network

```python
X = df.drop('diagnosis', axis = 1).values # Put this into numpy array format remember
y = df['diagnosis'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# Using a different scaler this time but still scaling our data
# This scaler converts all values to between 0 and 1
```

```python
X_train = scaler.fit_transform(X_train)
# We can fit and transform in the same step
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
# At a certain point in time, our validation loss begins to increase whilst training loss is still falling
# This tells us we are overfitting to our training dataset
# We can use early stopping to track the validation loss and stop the training once it starts overfitting
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)
# This means we are tracking the val loss, we want to minimise it, and we will wait 10 epochs after an increase is seen
# before we stop the NN running
```

```python
model = Sequential()

model.add(Dense(30,activation='relu')) # There are 30 features so good number of neurons to use

model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid')) # One final layer to decide malignant or benign

# Sigmoid function for binary classification

model.compile(loss='binary_crossentropy', optimizer='adam')
# Binary cross entropy as this is a binary classification problem
```

```python
model.fit(x=X_train, y=y_train, epochs=100, validation_data=(X_test, y_test), callbacks=[early_stop])
# Checking validation data after each epoch
# Using early stop to prevent additional loss, so we can use a number of epochs that is far too large
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
